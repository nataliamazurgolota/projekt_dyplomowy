```python
import pandas as pd
```

```python
df = pd.read_csv("/home/student/Pulpit/nypd.csv", low_memory=False)
```

```python
df2 = df.loc[(df["CONTRIBUTING FACTOR VEHICLE 1"] == "Unsafe Speed")]
df2
```

```python
df3 = df.loc[(df["CONTRIBUTING FACTOR VEHICLE 2"] == "Unsafe Speed")]
df3
```

```python
df4 = df.loc[(df["CONTRIBUTING FACTOR VEHICLE 3"] == "Unsafe Speed")]
df4
```

```python
df5 = df.loc[(df["CONTRIBUTING FACTOR VEHICLE 4"] == "Unsafe Speed")]
df5
```

```python
df6 = df.loc[(df["CONTRIBUTING FACTOR VEHICLE 5"] == "Unsafe Speed")]
df6
```

```python
data_raw = pd.concat([df2,df3,df4,df5,df6], ignore_index=True)
data_raw
```

```python
columns_list = ["ZIP CODE","LATITUDE","LONGITUDE","LOCATION","ON STREET NAME","CROSS STREET NAME","OFF STREET NAME","COLLISION_ID","VEHICLE TYPE CODE 1", "VEHICLE TYPE CODE 2","VEHICLE TYPE CODE 3","VEHICLE TYPE CODE 4","VEHICLE TYPE CODE 5"]
data = data_raw.drop(columns_list,axis=1)
data
```

```python
col_list = ["NUMBER OF PERSONS KILLED","NUMBER OF PEDESTRIANS KILLED","NUMBER OF CYCLIST KILLED","NUMBER OF MOTORIST KILLED"]
data["Total killed"]=data[col_list].sum(axis=1)
col_list2 = ["NUMBER OF PERSONS INJURED","NUMBER OF PEDESTRIANS INJURED","NUMBER OF CYCLIST INJURED","NUMBER OF MOTORIST INJURED"]
data["Total injured"]=data[col_list2].sum(axis=1)
data
```

```python
data2 = data[data['BOROUGH'].notna()]
data2
```

```python
columns_list = ["NUMBER OF PERSONS INJURED","NUMBER OF PERSONS KILLED","NUMBER OF PEDESTRIANS INJURED","NUMBER OF PEDESTRIANS KILLED","NUMBER OF CYCLIST INJURED","NUMBER OF CYCLIST KILLED","NUMBER OF MOTORIST INJURED","NUMBER OF MOTORIST KILLED"]
data_for_plot = data2.drop(columns_list,axis=1)
data_for_plot
```
