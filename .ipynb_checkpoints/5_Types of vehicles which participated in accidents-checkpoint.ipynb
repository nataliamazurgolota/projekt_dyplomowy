#### Import of libraries and uploading dataset from csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("/home/student/Pulpit/nypd.csv", low_memory=False)
```

#### Types of vehicles which participated in collisions are listed in five columns Vehicle type code 1-5. 
#### These types now will be counted and saved in five DataSeries, each for one column.

```python
results = []
results_renamed=[]
for i in range (1,6):
    column_name = "VEHICLE TYPE CODE "+str(i)
    results.append(pd.Series(raw_data.value_counts(column_name)).reset_index())
    results_renamed.append(results[i-1].rename(columns={column_name:"VEHICLE TYPE","count":"COUNT"}))
```

#### Merging all five Data Series into one using concat, then assuring that all types are typed in uppercase 
#### (e.g. two different types "TAXI" and "Taxi" now will be summed into one class "TAXI")

```python
merged_list_of_vehicles = pd.concat([results_renamed[0],
                                     results_renamed[1],
                                     results_renamed[2],
                                     results_renamed[3],
                                     results_renamed[4]], ignore_index=True)
```

```python
merged_list_of_vehicles['VEHICLE TYPE'] = merged_list_of_vehicles['VEHICLE TYPE'].str.upper()
```

#### Preparing data for plot, after merging repeated values will be summed. For plot 10 highest values will be used

```python
grouped_list_after_merging = pd.DataFrame(merged_list_of_vehicles.groupby(by='VEHICLE TYPE').sum())
sorted_list = grouped_list_after_merging.sort_values(by='COUNT',ascending=False)
top_10_of_vehicle_types = pd.DataFrame(sorted_list.head(10))
top_10_of_vehicle_types
```

```python
x = top_10_of_vehicle_types.index
y = top_10_of_vehicle_types["COUNT"]
fig = plt.figure(figsize = (10, 5))

plt.bar(x, y, color ='#C0C0C0', width = 0.4)
 
plt.xlabel("Type of vehicle")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of vehicles")
plt.ticklabel_format(style='plain', axis="y")
plt.title("Top 10 of vehicle types participating in NYC accidents")
plt.show()
```

### Conclusions
#### Most common type of vehicle participated in collisions in NYPD between 2012 and 2019, was passenger vehicle. 
#### Worth noticing is fact, that SUVs and station wagons are connected into one group but under two different names.
